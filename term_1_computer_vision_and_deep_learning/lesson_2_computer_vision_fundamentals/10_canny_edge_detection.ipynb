```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np
import cv2
```

```python
image = mpimg.imread('exit-ramp.jpg')
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
image
```

```python
# https://en.wikipedia.org/wiki/Grayscale
# https://www.tutorialspoint.com/dip/grayscale_to_rgb_conversion.htm
# https://docs.opencv.org/2.4/modules/imgproc/doc/miscellaneous_transformations.html#cvtcolor
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
# https://matplotlib.org/tutorials/colors/colormaps.html
# https://matplotlib.org/examples/color/colormaps_reference.html
plt.imshow(gray, cmap='gray')
```

```python
gray.shape
```

```python
gray
```

```python
# https://docs.opencv.org/2.4/modules/imgproc/doc/filtering.html?highlight=gaussianblur#gaussianblur
kernel_size = 3
blur_gray = cv2.GaussianBlur(gray, (kernel_size, kernel_size), 0)
```

```python
plt.imshow(blur_gray, cmap='gray')
```

```python
blur_gray
```

```python
# https://docs.opencv.org/2.4/doc/tutorials/imgproc/imgtrans/canny_detector/canny_detector.html#steps
# https://docs.opencv.org/2.4/modules/imgproc/doc/feature_detection.html?highlight=canny#canny

low_threshold = 1
high_threshold = 10

edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
```

```python
edges
```

```python
plt.imshow(edges, cmap='Greys_r')
```

```python
# low_threshold = 1
# high_threshold = 10

low_threshold = 50
high_threshold = 200

edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
```

```python
plt.imshow(edges, cmap='Greys_r')
```

```python
# low_threshold = 1
# high_threshold = 10

low_threshold = 60
high_threshold = 150

edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
```

```python
plt.imshow(edges, cmap='Greys_r')
```

```python
mpimg.imsave('actual_edges_exit_ramp.jpg', edges, cmap='Greys_r')
```
