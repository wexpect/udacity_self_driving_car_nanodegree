```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
nx = 8
ny = 6
```

```python
fname = 'calibration_test.png'
img = cv2.imread(fname)
plt.imshow(img)
```

```python
img.shape[1::-1]
```

```python
img.shape, img
```

```python
#  If you are reading in an image using mpimg.imread() this will read in an RGB image and you should convert to grayscale using cv2.COLOR_RGB2GRAY, but if you are using cv2.imread() or the glob API, as happens in this video example, this will read in a BGR image and you should convert to grayscale using cv2.COLOR_BGR2GRAY.
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
gray.shape[::-1]
```

```python
gray.shape
```

```python
plt.imshow(gray, cmap='Greys_r')
```

```python
ret, corners = cv2.findChessboardCorners(gray, (nx, ny), None)
ret
```

```python
corners.shape
```

```python
corners
```

```python
if ret == True:
    cv2.drawChessboardCorners(img, (nx, ny), corners, ret)
    plt.imshow(img)
```
