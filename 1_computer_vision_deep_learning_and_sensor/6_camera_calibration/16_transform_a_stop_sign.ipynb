```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('stop_sign.png')
img.shape
```

```python
%matplotlib qt
plt.imshow(img)
```

```python
%matplotlib inline
plt.imshow(img)
plt.plot(882, 340, '.')
plt.plot(905, 466, '.')
plt.plot(551, 368, '.')
plt.plot(560, 233, '.')
```

```python
def warp(img):
    img_size = (img.shape[1], img.shape[0])
    
    src = np.float32([
        [882, 340],
        [905, 466],
        [551, 368],
        [560, 233]        
    ])
    
    dst = np.float32([
        [870, 240],
        [870, 370],
        [520, 370],
        [520, 240]
                
#         [600, 500],
#         [600, 600],
#         [400, 600],
#         [400, 500]   
                
#         [900, 50],
#         [900, 1000],
#         [50, 1000],
#         [50, 50]        
    ])
    
    M = cv2.getPerspectiveTransform(src, dst)
    
    M_inv = cv2.getPerspectiveTransform(dst, src)
    
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    
    return warped
```

```python
# [870, 240],
# [870, 370],
# [520, 370],
# [520, 240]

warped_img = warp(img)

%matplotlib inline
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
ax1.set_title('Source Img')
ax1.imshow(img)
ax2.set_title('Warped Img')
ax2.imshow(warped_img)
```
